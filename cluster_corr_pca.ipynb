# PCA TSNE
PCA(Principal Component Analysis)，即主成分分析方法，是一种使用最广泛的数据降维算法。PCA的主要思想是将n维特征映射到k维上，这k维是全新的正交特征也被称为主成分，是在原有n维特征的基础上重新构造出来的k维特征。

t-SNE是目前来说效果最好的数据降维与可视化方法，但是它的缺点也很明显，比如：占内存大，运行时间长。但是，当我们想要对高维数据进行分类，又不清楚这个数据集有没有很好的可分性，可以通过t-SNE投影到2维或者3维的空间中观察一下。

```python
df = pd.read_csv(r"C:\Users\lydz\Desktop\FLNC\merge.RPM", sep = '\t', index_col= 0)
print(df)
```

```python
pca_data_1 = df[df.mean(axis =1) > 1 ]# 筛选样本平均表达量大于1的表达矩阵。[mean(axis=1):按行求均值；axis=0:按列执行。]
pca_data_2 = df[df.mean(axis =1) > 1 ]**.5# 筛选样本平均表达量大于1的表达矩阵并对每个表达量值进行开方
pca_data_3 = np.log2(df[df.mean(axis =1) > 1 ]+1)#筛选样本平均表达量大于1的表达矩阵并对每个表达量值进行+1后取log2对数
```

```python
pca = decomposition.PCA(n_components=2)
#简写函数，把数据降到两个维度
pca_data_1.T
```

```python
pca.fit_transform(pca_data_1.T)
#newX=pca.fit_transform（X）:用X来训练PCA模型，同时返回降维后的数据，newX就是降维后的数据
#pca_data_1.T：T 转置了一下数据
```

```python
pca_1 = pd.DataFrame(pca.fit_transform(pca_data_1.T), columns=['PC1', 'PC2'] , index = ['DAP10-1','DAP10-2', 'DAP10-3', 'DAP18-1', 'DAP18-2', 'DAP18-3', 'DAP26-1', 'DAP26-2', 'DAP26-3', 'DAP34-1', 'DAP34-2', 'DAP34-3']).reset_index()
pca_2 = pd.DataFrame(pca.fit_transform(pca_data_2.T), columns=['PC1', 'PC2'] , index = ['DAP10-1','DAP10-2', 'DAP10-3', 'DAP18-1', 'DAP18-2', 'DAP18-3', 'DAP26-1', 'DAP26-2', 'DAP26-3', 'DAP34-1', 'DAP34-2', 'DAP34-3']).reset_index()
pca_3 = pd.DataFrame(pca.fit_transform(pca_data_3.T), columns=['PC1', 'PC2'] , index = ['DAP10-1','DAP10-2', 'DAP10-3', 'DAP18-1', 'DAP18-2', 'DAP18-3', 'DAP26-1', 'DAP26-2', 'DAP26-3', 'DAP34-1', 'DAP34-2', 'DAP34-3']).reset_index()
```

# 聚类

k均值聚类算法（k-means clustering algorithm）是一种迭代求解的聚类分析算法，其步骤是，预将数据分为K组，则随机选取K个对象作为初始的聚类中心，然后计算每个对象与各个种子聚类中心之间的距离，把每个对象分配给距离它最近的聚类中心。聚类中心以及分配给它们的对象就代表一个聚类。





BIRCH的全称是利用层次方法的平衡迭代规约和聚类


```python
from scipy.stats import zscore
from sklearn.cluster import KMeans
from sklearn.cluster import Birch
```

```python
kmeans = KMeans(n_clusters=11)
#####简写函数，n_cluster=5，预先分成5组的意思？
birch = Birch(n_clusters=11)
```

```python
matrix = np.log2(df[df.mean(axis = 1) != 0]+1)#筛选样本平均表达量不等于0的表达矩阵并对每个表达量值+1后取log2对数
k_cluster = kmeans.fit(matrix)
#KMeans().fit(data),data里的数据为矩阵数据，矩阵为（数据点）x（数据类型）的规格这里为n x 2;聚类根据两个类型数据的n个观测值进行聚类
matrix['mean'] = matrix.mean(axis=1)#求均值,这样不对，把label算进去了？
matrix['label'] = k_cluster.labels_

#label是分组的意思？

matrix.sort_values(["label", 'mean'], inplace=True)
matrix
#sort函数：按照label和mean进行排序，inplace=True:是否用排序后的矩阵替换原矩阵，默认为false
```

```python
plt.figure(figsize=(4,10))
sns.heatmap(matrix.iloc[:, :4], cmap="RdBu_r", vmax = 8)
#cmap:设置颜色带的色系
```

Z-Score的主要目的就是将不同量级的数据统一转化为同一个量级，统一用计算出的Z-Score值衡量，以保证数据之间的可比性。
所以是将每一个特征进行标准化。

```python
df_none = df[df.mean(axis = 1) != 0]#去掉0值
matrix_df = pd.DataFrame(zscore(df_none, axis =1), columns=df_none.columns, index = df_none.index)
matrix_df
```

```python
k_cluster_1 = kmeans.fit(matrix_df)
#KMeans().fit(data),data里的数据为矩阵数据，矩阵为（数据点）x（数据类型）的规格这里为n x 2;聚类根据两个类型数据的n个观测值进行聚类
matrix_df['label'] = k_cluster_1.labels_
matrix_df
matrix_df.sort_values("label", inplace=True)
matrix_df
```

```python
plt.figure(figsize=(4,10))
sns.heatmap(matrix_df.iloc[:, :4], cmap="RdBu_r", vmax =3, vmin = -3)
```

```python
df_none = df[df.mean(axis = 1) != 0]
matrix_df = pd.DataFrame(zscore(df_none, axis =1), columns=df_none.columns, index = df_none.index)
k_cluster_2 = birch.fit(matrix_df)
matrix_df['label_1'] = k_cluster_2.labels_
matrix_df.sort_values("label_1", inplace=True)
```

```python
plt.figure(figsize=(4,10))
sns.heatmap(matrix_df.iloc[:, :4], cmap="RdBu_r", vmax =3, vmin = -3)
```

```python
df_none.reset_index().values
#数据清洗时，会将带空值的行删除，此时DataFrame或Series类型的数据不再是连续的索引，可以使用reset_index()重置索引
#df.column.values   以array形式返回指定column的所有取值
```

```python
final = []
one = [3.410,3.370,3.372,4.601]
for nu in df_none.reset_index().values[:10000]:#为什么要取前10000行？
    corr, pv = stats.spearmanr([1.000000,0.928875,0.917173,0.816711,0.807714,0.776173,0.731290,0.731930,0.731274,0.686718,0.697040,0.701918], nu[1:])
    final.append(['MSTRG.10000',nu[0], corr, pv ])
#求了基因间两两的corr和pv；
```

```python
final_Df = pd.DataFrame(final, columns=['ge1', 'ge2', 'corr', 'pv'])
print(final_Df)
```

```python
ge_corr = final_Df[final_Df['pv']<0.05]#筛选了PV<0.05的列
```

```python
ge_corr_plus = ge_corr[ge_corr['corr']>0]
ge_corr_minus = ge_corr[ge_corr['corr']<0]
```

```python
one_gene_plus = pd.DataFrame(df_none.loc[ge_corr_plus['ge2']].stack()).reset_index()
one_gene_minus = pd.DataFrame(df_none.loc[ge_corr_minus['ge2']].stack()).reset_index()
```
